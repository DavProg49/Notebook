Exercice avec les survivants du titanic (manipulation librairie pandas)

```
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')

#apercu des données
titanic.head()
```

```
# un coup d'oeil aux âges, unique va renvoyer les valeurs uniques présentes 
# dans la structure de données Pandas.
titanic.age.unique()
```

```
#fonction describe pour donner des statistiques (moyenne, maxi, mini, etc...) 
#sur les données de chaque colonne
titanic.describe(include="all") # all, inclus les colomnes non numériques
```

```
titanic.age.head(10)
```

```
# remplacer les valeurs NaN par 0
titanic.fillna(value={"age": 0}).age.head(10)
```

```
# on remplace par les valeurs précedentes 
titanic.fillna(method="pad").age.head(10)
```

```
# suppression des axes colonnes ou lignes ( par défaut les lignes )
titanic.dropna().head(10)
```

```
titanic.dropna(axis="columns").head()
```

```
#renommer une colonne
titanic.rename(columns={"sex":"sexe"})
```

```
# application d'une fonction à tous les index 
f = lambda x: x+1
titanic.rename(index=f)
```

```
# suppression des axes, ici la ligne 0
titanic.drop(0)
```

```
# suppression de la colonne "age"
titanic.drop(columns=["age"])
```

```
# Tableaux dynamiques
# table de pivot sur la répartition des survivants 
# en fonction de leurs sexes et type de billet
titanic.pivot_table('survived', index='sex', columns='class')
```

```
# nombre total de survivant par sexes et type de billet
titanic.pivot_table('survived', index='sex', columns='class', aggfunc="sum")
```

```
# on peut recouper en catégorie ici 0-18 ans et 18-80 ans
# le résultat est un dataframe
titanic.dropna(inplace=True)
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', ['sex', age], 'class')
```
